## Declare feature vector and target variable

```python
X = tr.drop(['target'], axis=1)
y = tr['target']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

## Model fitting and feature selection

```python
sel = SelectFromModel(RandomForestClassifier(n_estimators=100))
sel.fit(X_train,y_train)
```

## Seeing which features are important

```python
sel.get_support()
```

```python
selected_feat= X_train.columns[(sel.get_support())]
print(len(selected_feat))
print(selected_feat)
```

```python
X_train[selected_feat]
```

```python
sel.estimator_
```

```python
#sel.estimator_.predict(X_test[selected_feat])
model_with_selected_feats = RandomForestClassifier()
model_with_selected_feats.fit(X_train[selected_feat], y_train)

preds = model_with_selected_feats.predict(X_test[selected_feat])
np.mean(preds == y_test)
```
